```python
# Cell 1: imports
import time
import src.scanpy as sc
from src.scanpy.plotting._dotplot import DotPlot
from src.scanpy.plotting._matrixplot import MatrixPlot
from src.scanpy.plotting._stacked_violin import StackedViolin
```

```python
# Cell 2: load data
adata = sc.datasets.pbmc3k_processed()
# pick a handful of genes and a grouping key that exists
var_names = adata.var_names[:20]
groupby   = "louvain"   # or any categorical obs field in this adata
```

```python
# Cell 3: time DotPlot
start = time.time()
dp = DotPlot(adata, var_names, groupby)
dp.add_totals(sort="descending", ascending=False)
print("DotPlot total time", time.time() - start, "seconds")
# quick sanity check
print(" dot size df shape", dp.dot_size_df.shape)
print(" dot color df shape", dp.dot_color_df.shape)
```

```python
# Cell 4: time MatrixPlot
start = time.time()
mp = MatrixPlot(adata, var_names, groupby)
mp.add_totals(sort="descending", ascending=False)
print("MatrixPlot total time", time.time() - start, "seconds")
# quick sanity check
print(" values df shape", mp.values_df.shape)
```

```python
# Cell 5: time StackedViolin
start = time.time()
sv = StackedViolin(adata, var_names, groupby)
# you can optionally call add_totals if you use it in your workflows
sv.add_totals(sort="descending", ascending=False)
print("StackedViolin total time:", time.time() - start, "s")
print(" obs_tidy shape:", sv.obs_tidy.shape)
```
